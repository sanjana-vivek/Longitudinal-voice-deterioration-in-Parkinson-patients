```python
!mkdir -p ~/.kaggle
!mv kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json


!kaggle datasets download -d cycoool29/movement-disorders-voice


!unzip -o movement-disorders-voice.zip -d ./data

import os
import zipfile

zip_path = "movement-disorders-voice.zip"
extract_dir = "data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Files extracted successfully!")


EXTRACT_DIR = "data/data/Raw Audio"
CLASSES = ['Parkinson', 'Alzheimer', 'Healthy']
N_MFCC = 40
FEATURES_FILE = "wavelet_mfcc_phonation_features.csv"


import numpy as np
import matplotlib.pyplot as plt
import pywt
import librosa
import pandas as pd
from joblib import Parallel, delayed


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier

import seaborn as sns


def load_audio(file_path, sr=8000):
    signal, fs = librosa.load(file_path, sr=sr, mono=True)
    signal = signal / np.max(np.abs(signal))
    return signal, fs


def wavelet_transform(signal, fs):

    scales = np.linspace(1, 64, 32, dtype=int)
    coefficients, frequencies = pywt.cwt(signal, scales, 'morl', sampling_period=1/fs)
    return coefficients, scales


def extract_wavelet_features(coefficients, scales):
    energy_per_scale = np.sum(np.abs(coefficients)**2, axis=1)
    tremor_energy = np.sum(energy_per_scale[scales < 20])
    breathiness_energy = np.sum(energy_per_scale[scales > 40])
    prob_dist = energy_per_scale / np.sum(energy_per_scale)
    wavelet_entropy = -np.sum(prob_dist * np.log2(prob_dist + 1e-12))
    return {
        "tremor_energy": tremor_energy,
        "breathiness_energy": breathiness_energy,
        "wavelet_entropy": wavelet_entropy
    }


def extract_mfcc_features(signal, fs, n_mfcc=40):
    mfcc = librosa.feature.mfcc(y=signal, sr=fs, n_mfcc=n_mfcc)
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)
    features = {}
    for i in range(n_mfcc):
        features[f"mfcc_{i+1}_mean"] = mfcc_mean[i]
        features[f"mfcc_{i+1}_std"] = mfcc_std[i]
    return features


def extract_phonation_features(signal, fs):
    features = {}


    try:
        f0 = librosa.yin(signal, fmin=50, fmax=500, sr=fs)
        f0 = f0[f0 > 0]
        if len(f0) > 1:
            jitter = np.mean(np.abs(np.diff(f0)) / f0[:-1])
        else:
            jitter = 0.0
    except:
        jitter = 0.0


    frame_length = 1024
    hop_length = 512
    amplitude_envelope = np.array([
        max(signal[i:i+frame_length]) if i+frame_length < len(signal) else max(signal[i:])
        for i in range(0, len(signal), hop_length)
    ])
    if len(amplitude_envelope) > 1:
        shimmer = np.mean(np.abs(np.diff(amplitude_envelope)) / (amplitude_envelope[:-1] + 1e-12))
    else:
        shimmer = 0.0


    try:
        S = librosa.stft(signal)
        harmonic, percussive = librosa.decompose.hpss(S)
        hnr = 10 * np.log10(np.sum(np.abs(harmonic)**2) / (np.sum(np.abs(percussive)**2) + 1e-12))
    except:
        hnr = 0.0

    features["jitter"] = jitter
    features["shimmer"] = shimmer
    features["hnr"] = hnr

    return features


def process_file(path, cls, sr=8000):
    try:
        signal, fs = load_audio(path, sr=sr)


        coeffs, scales = wavelet_transform(signal, fs)
        feats_wavelet = extract_wavelet_features(coeffs, scales)


        feats_mfcc = extract_mfcc_features(signal, fs, n_mfcc=N_MFCC)


        feats_phonation = extract_phonation_features(signal, fs)


        feats = {**feats_wavelet, **feats_mfcc, **feats_phonation}
        feats["filename"] = os.path.basename(path)
        feats["label"] = cls
        return feats
    except Exception as e:
        print(f"Error processing {path}: {e}")
        return None


def process_dataset(folder_path, sr=8000):
    tasks = []
    for cls in CLASSES:
        class_path = os.path.join(folder_path, cls)
        if os.path.exists(class_path):
            files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith(".wav")]
            tasks.extend((path, cls) for path in files)
    results = Parallel(n_jobs=-1)(delayed(process_file)(path, cls, sr) for path, cls in tasks)
    return pd.DataFrame([r for r in results if r is not None])


if os.path.exists(FEATURES_FILE):
    df_features = pd.read_csv(FEATURES_FILE)
    print("Loaded cached features!")
else:
    df_features = process_dataset(EXTRACT_DIR)
    df_features.to_csv(FEATURES_FILE, index=False)
    print("Features extracted and cached.")

display(df_features.head())


plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="tremor_energy", data=df_features)
plt.title("Tremor Energy Distribution by Class")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="breathiness_energy", data=df_features)
plt.title("Breathiness Energy Distribution by Class")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="jitter", data=df_features)
plt.title("Jitter Distribution by Class")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="shimmer", data=df_features)
plt.title("Shimmer Distribution by Class")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="hnr", data=df_features)
plt.title("HNR Distribution by Class")
plt.show()


X = df_features.drop(columns=["filename", "label"])
y = df_features["label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=500, random_state=42),
    "SVM": SVC(kernel="rbf", probability=True, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Extra Trees": ExtraTreesClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, random_state=42)
}


for name, model in models.items():
    print("="*60)
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Results for {name}:")
    print(classification_report(y_test, y_pred, digits=4))
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=y.unique(), yticklabels=y.unique()
    )
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

```
